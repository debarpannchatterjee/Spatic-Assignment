## Importing Necessary Libraries

```python
import pandas as pd
import numpy as np
from geopy.distance import geodesic
from Levenshtein import distance as levenshtein_distance
from tqdm import tqdm
```

## Checking duplicates and shape

```python
df = pd.read_csv('assignment_data.csv')
print('checking duplicates and shape')
df["name"].nunique(), df.shape

```

## Defining Necessary Functions

```python
#function to calculate geographic distance
def dist(entry1, entry2):
    dist = geodesic((entry1['latitude'], entry1['longitude']), (entry2['latitude'], entry2['longitude'])).m
    if(dist<=200):
        return True
#function to calculate levenshtein distance    
def name_dist(entry1, entry2):
    name_distance = levenshtein_distance(entry1['name'], entry2['name'])
    if(name_distance < 5):
        return True
    
```

## Iterating through the df twice to compare each record and find a similar pair

```python
similar_entries = []

for i in tqdm(range(len(df))):
    for j in range(i+1, len(df)):
        if name_dist(df.loc[i], df.loc[j]):
            if dist(df.loc[i], df.loc[j]):
                similar_entries.append((i, j))
            
print(similar_entries)
```

## Creating a new column and exporting as CSV file

```python
# Mark the similar entries in the dataframe
df['is_similar'] = False
for i, j in similar_entries:
    df.at[i, 'is_similar'] = True
    df.at[j, 'is_similar'] = True

# Save the updated dataframe as a CSV file
df.to_csv('similar_entries.csv', index=False)
```
